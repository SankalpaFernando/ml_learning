```python
# Absolute value conversion
tensor1 = torch.Tensor([[1,-2,3],[-3,-4,1]])
tensor2 = torch.Tensor([[3,4,-2],[0,3,4]])
```

```python
tensor1.abs_()
```

```python
tensor2.abs_()
```

```python
# Matrix Addition
add_matrix = tensor1 + tensor2
add_matrix
```

```python
add_matrix = torch.add(tensor1,tensor2)
add_matrix
```

```python
# Elementwise Divison
div_matrix = torch.div(tensor1,tensor1+0.3) # dividing each element in the tensor by the value itself + 0.3
div_matrix
```

```python
#Elementwise Multiplication
mul_matrix = torch.mul(tensor1,tensor1)
mul_matrix
```

```python
#Clamp Operation - used to clamp the tensor values between a max and a min value
print(tensor1)
tensor1.clamp(min=0,max=2)
```

```python
# Dot Product
t1 = torch.Tensor([2,3,4])
t2 = torch.Tensor([3,4,1])

torch.dot(t1,t2)
```

```python
# Multiply different dimensions

matrix = torch.Tensor([[1,2,3],[3,4,5]])
vector = torch.Tensor([1,2,2])

torch.mv(matrix,vector)
```

```python
# Matrix Multiplication

t1 = torch.Tensor([[1,2,3],[3,4,5]]) # 2 x 3
t2 = torch.Tensor([[1,2],[3,4],[5,6]]) # 3 x 2

torch.mm(t1,t2)
```

```python
# Finding the largest dimension in a particular dimension
torch.argmax(t1,dim=1)
```

```python
torch.argmin(t1,dim=1)
```
