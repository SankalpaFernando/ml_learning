```python
import numpy as np
import torch
```

```python
tensor = torch.rand(4,3)
tensor
```

```python
type(tensor)
```

```python
# Tensor to Numpy
numpy_from_tensor = tensor.numpy()
numpy_from_tensor
```

```python
numpy_array = np.array([[1,3,2,3],[5,6,7,3]])
numpy_array
```

```python
# Numpy to Tensor
tensor = torch.from_numpy(numpy_array)
tensor
```

```python
# Even Once converted both share the memory, thus doing any changes on one, will reflect on other
```

```python
# To create a tensor without sharing memory only values
tensor_copy = torch.tensor(numpy_array)
tensor_copy
```
